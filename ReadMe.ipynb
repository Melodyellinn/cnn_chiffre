### **Here is a small ReadMe, to explain what is in the GitHub "cnn_chiffre"**

=>>> https://github.com/Melodyellinn/cnn_chiffre

**notebook_model** explains the whole process :  
import data  
preprocessing  
creat model  

**notebook_predict** repeats the process to verify that the model is loaded and working.  

**model.h5** is our model saved and loaded by our application.  

**app000.py** is our file that contains the commands to generate our application that Streamlit will read.  

**Procfile** is a mechanism to declare which commands are executed by the dynos of the application.  

**requirement.txt**  
to get the package versions that Streamlit should load.  

**removed.txt**  
Remove opencv-python, that did not work to deploy it on Streamlit.  

**setup.sh** Defines the Bourne or Korn shell environment variables needed to build an InfoCrafter database.  

Go on Streamlit app !  

**Here**  

=>>> https://share.streamlit.io/melodyellinn/cnn_chiffre/main/app000.py
